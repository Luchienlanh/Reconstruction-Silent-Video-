```python
import whisper
import torch
import json
import os
import glob

# Cấu hình thiết bị
device = "cuda" if torch.cuda.is_available() else "cpu"
model = whisper.load_model("turbo", device=device)

def transcribe_audio(audio_path, target_folder):
    """Hàm xử lý và lưu kết quả vào thư mục được chỉ định"""
    # 1. Chạy Whisper
    results = model.transcribe(audio_path, verbose=False, word_timestamps=True)
    word_res = []
    for segment in results['segments']:
        for word in segment['words']:
            word_res.append({
                'word': word['word'].strip(),
                'start': round(word['start'], 3),
                'end': round(word['end'], 3),
            })

    # 2. Xử lý đường dẫn lưu file
    file_name = os.path.basename(audio_path) # Lấy tên file: "video1.mp4"
    json_name = os.path.splitext(file_name)[0] + ".json" # Đổi tên: "video1.json"
    json_path = os.path.join(target_folder, json_name) # Ghép vào đường dẫn đích

    # 3. Ghi file
    with open(json_path, 'w', encoding='utf-8') as f:
        json.dump(word_res, f, ensure_ascii=False, indent=4)

    return json_path

if __name__ == "__main__":
    video_dir = r"/content/drive/MyDrive/vtv24_videos"
    output_dir = r"/content/drive/MyDrive/JsonFile"

    # Tạo thư mục đích nếu chưa có
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        print(f"📂 Đã tạo thư mục lưu trữ: {output_dir}")

    mp4_files = glob.glob(os.path.join(video_dir, "*.mp4"))
    print(f"🔍 Tìm thấy {len(mp4_files)} video. Bắt đầu xử lý...")

    for video_goc in mp4_files:
        # Kiểm tra file đã tồn tại trong thư mục JsonFile chưa
        file_name = os.path.basename(video_goc)
        json_file_name = os.path.splitext(file_name)[0] + ".json"
        json_check_path = os.path.join(output_dir, json_file_name)

        if os.path.exists(json_check_path):
            print(f"⏩ Bỏ qua (Đã có file JSON): {json_file_name}")
            continue

        print(f"\n🎥 Đang xử lý: {file_name}")
        try:
            # Truyền thêm output_dir vào hàm
            json_file = transcribe_audio(video_goc, output_dir)
            print(f"✅ Đã lưu: {os.path.basename(json_file)}")
        except Exception as e:
            print(f"❌ Lỗi khi xử lý {file_name}: {e}")
```
